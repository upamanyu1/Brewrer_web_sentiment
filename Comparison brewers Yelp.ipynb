```python
import requests
from bs4 import BeautifulSoup 
```

```python
url_1 = 'https://www.yelp.com/biz/yards-brewing-company-philadelphia-3'
response_1 = requests.get(url_1)
if response_1.status_code == 200:
    print("Success")
else:
    print("Failure")
```

```python
url_2 = 'https://www.yelp.com/biz/philadelphia-brewing-company-philadelphia?osq=Philadelphia+Brewing+Company'
response_2 = requests.get(url_2)
if response_2.status_code == 200:
    print("Success")
else:
    print("Failure")
```

```python
response_1_html = BeautifulSoup(response_1.text,"html.parser")
```

```python
response_1_html
```

```python
response_2_html = BeautifulSoup(response_2.text,"html.parser")
```

```python
response_2_html
```

```python
res_1=response_1_html.findAll("span", {"class": "raw__373c0__3rcx7"})
```

```python
res_2=response_2_html.findAll("span", {"class": "raw__373c0__3rcx7"})
```

```python
l1=len(res_1)
```

```python
l1
```

```python
l2=len(res_2)
```

```python
l2
```

```python
review1=[]
for i in range(l1):
    review1.append(response_1_html.findAll("span", {"class": "raw__373c0__3rcx7"})[i].text)
    
```

```python
review1
```

```python
review2=[]
for i in range(l2):
    review2.append(response_2_html.findAll("span", {"class": "raw__373c0__3rcx7"})[i].text)
    
```

```python
review2
```

```python
type(review2)
```

```python
str1 = " " 
s1=str1.join(review1)
```

```python
str2 = " " 
s2=str2.join(review2)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline

wordcloud1 = WordCloud(stopwords=STOPWORDS,background_color='white',width=6000,height=6000).generate(s1)
wordcloud2 = WordCloud(stopwords=STOPWORDS,background_color='white',width=6000,height=6000).generate(s2)

fig, axs = plt.subplots(ncols=2, sharex=True, figsize=(20, 25))
axs[0].set_title('customer review of Yard Brewing Co')
axs[0].imshow(wordcloud1)

axs[1].set_title('customer review of philadelphia brewing co')
axs[1].imshow(wordcloud2)
plt.axis('off')
plt.show()
```
